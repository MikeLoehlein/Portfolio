```python
'''Simple Programming Loop to Calculate first 100 FizzBuzz Numbers'''

fizbuzz = []
fizz = []
buzz = []

i=0
number = 1
while i < 100:
    if (number % 3 == 0) & (number % 5 == 0):
        fizbuzz.append(number)
        i += 1
    elif (number % 3 == 0):
        fizz.append(number)
    elif (number % 5 == 0):
        buzz.append(number)
        
    number += 1
    
print(len(fizbuzz))
print(fizbuzz)
print()

print(buzz)
print()

print(fizz)
print()
```
